## Project Description
This project analyzes student performance data to identify factors that influence academic success. The analysis focuses on gender and parental education level.

## Tools Used
- Python
- Pandas
- Matplotlib

```python
import pandas as pd
import numpy as np

# إنشاء بيانات طلاب بشكل مباشر
np.random.seed(42)

data = {
    "gender": np.random.choice(["male", "female"], 100),
    "parental level of education": np.random.choice([
        "high school", "bachelor", "master"
    ], 100),
    "math score": np.random.randint(50, 100, 100),
    "reading score": np.random.randint(50, 100, 100),
    "writing score": np.random.randint(50, 100, 100)
}

df = pd.DataFrame(data)

df.head()
```
